```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

# グラフの描画

```python
data = np.arange(0,10)
plt.plot(data)
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
plt.plot(np.random.randn(10).cumsum())

ax2 = fig.add_subplot(2,2,2)
plt.plot(np.random.randn(10).cumsum(), 'r--')

ax3 = fig.add_subplot(2,2,3)
plt.plot(np.random.randn(10).cumsum(), 'go--')

ax4 = fig.add_subplot(2,2,4)
plt.plot(np.random.randn(10).cumsum(), 'k+-')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_xticks(np.arange(0,10,1))
ax.set_yticks(np.arange(-5,5,1))
ax.set_title('title')
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
ax.plot(np.random.randn(10).cumsum())
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_xticks(np.arange(0,10,1))
ax.set_yticks(np.arange(-10,10,1))
ax.set_title('title')
ax.set_xlabel('xlabel')
ax.set_ylabel('ylabel')
ax.plot(np.random.randn(10).cumsum(), label='label1')
ax.plot(np.random.randn(10).cumsum(), label='label2')
ax.plot(np.random.randn(10).cumsum(), label='label3')
ax.legend()
```
